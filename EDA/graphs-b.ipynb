In here, graph various relationships in the data

```python
import pandas as pd 
import matplotlib.pyplot as plt

# import and load data 
mm_data = pd.read_csv("../Data/DEV _ March Madness.csv")
```

```python
# graph the distribution of active coaching length

plt.figure(figsize=(10, 6))
plt.hist(mm_data['Active Coaching Length Index'], bins=30, edgecolor='black', alpha=0.7)
plt.xlabel('Active Coaching Length (years)')
plt.ylabel('Frequency')
plt.title('Distribution of Active Coaching Length')
plt.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
print((mm_data[mm_data["Active Coaching Length Index"] > 25]["Full Team Name"]).unique())
```

```python
print(mm_data[mm_data["Full Team Name"] == "Duke Blue Devils"][["Current Coach", "Season"]])
# The data does not contain info about a team's coach at the time of the entry, only the current coach, which is not very helpful
```

```python
# Count the frequency of each unique value in EffectiveHeight
# Round to nearest whole number
mm_data['EffectiveHeight_Rounded'] = mm_data['EffectiveHeight'].round()

# Count the frequency of each rounded value
height_counts = mm_data['EffectiveHeight_Rounded'].value_counts().sort_index()

# Create a pie chart
plt.figure(figsize=(10, 8))
plt.pie(height_counts, labels=height_counts.index, autopct='%1.1f%%', startangle=90)
plt.title('Distribution of Effective Height (Rounded)')
plt.axis('equal')
plt.tight_layout()
plt.show()
```
